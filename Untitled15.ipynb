```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Game Variables
screen_width = 400
screen_height = 600
gravity = 0.25
bird_movement = 0
game_active = True

# Set up the screen
screen = pygame.display.set_mode((screen_width, screen_height))

# Load images
bg_surface = pygame.image.load('assets/background-day.png').convert()
bg_surface = pygame.transform.scale2x(bg_surface)

bird_surface = pygame.image.load('assets/bluebird-midflap.png').convert_alpha()
bird_surface = pygame.transform.scale2x(bird_surface)
bird_rect = bird_surface.get_rect(center=(100, screen_height // 2))

pipe_surface = pygame.image.load('assets/pipe-green.png').convert()
pipe_surface = pygame.transform.scale2x(pipe_surface)
pipe_list = []

# Time
pipe_spawn = pygame.USEREVENT
pygame.time.set_timer(pipe_spawn, 1200)
game_over_font = pygame.font.Font('04B_19.ttf', 40)

def draw_background():
    screen.blit(bg_surface, (0, 0))

def draw_bird():
    screen.blit(bird_surface, bird_rect)

def create_pipe():
    random_pipe_pos = random.choice(pipe_height)
    bottom_pipe = pipe_surface.get_rect(midtop=(500, random_pipe_pos))
    top_pipe = pipe_surface.get_rect(midbottom=(500, random_pipe_pos - 300))
    return bottom_pipe, top_pipe

def draw_pipes(pipes):
    for pipe in pipes:
        if pipe.bottom >= 600:
            screen.blit(pipe_surface, pipe)
        else:
            flip_pipe = pygame.transform.flip(pipe_surface, False, True)
            screen.blit(flip_pipe, pipe)

def check_collision(pipes):
    for pipe in pipes:
        if bird_rect.colliderect(pipe):
            return False
    if bird_rect.top <= -100 or bird_rect.bottom >= 900:
        return False
    return True

def rotate_bird(bird):
    new_bird = pygame.transform.rotozoom(bird, -bird_movement * 3, 1)
    return new_bird

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE and game_active:
                bird_movement = 0
                bird_movement -= 6
            if event.key == pygame.K_SPACE and not game_active:
                game_active = True
                pipe_list.clear()
                bird_rect.center = (100, screen_height // 2)
                bird_movement = 0
        if event.type == pipe_spawn:
            pipe_list.extend(create_pipe())

    draw_background()

    if game_active:
        # Bird
        bird_movement += gravity
        rotated_bird = rotate_bird(bird_surface)
        bird_rect.centery += bird_movement
        draw_bird()
        game_active = check_collision(pipe_list)

        # Pipes
        draw_pipes(pipe_list)

    pygame.display.update()
    pygame.time.Clock().tick(120)

pygame.quit()











```

```python
def find_average_followers(country):
    B = []
    f = open("plaintext.txt", "r")
    for line in f:
        line = line.strip()
        line1 = line.split(";")
        B.append(line1)
    f.close()

    B1 = B[1:]
    A = {}

    for x in B1:
        if len(x) > 5:
            if "+" in x[5]:
                x1 = x[5].split("+")
                for x5 in x1:
                    if x5 not in A:
                        A[x5] = [float(x[3])]
                    else:
                        A[x5].append(float(x[3]))
            else:
                x5 = x[5]
                if x5 not in A:
                    A[x5] = [float(x[3])]
                else:
                    A[x5].append(float(x[3]))

    for x in A:
        A[x] = sum(A[x]) / len(A[x])

    if country in A:
        average_followers = A[country]
        print(f"The average number of followers for {country} is {average_followers}")
    else:
        print(f"No data available for {country}")


# Example usage
find_average_followers("United States")
```

```python
import requests
import pandas as pd
from collections import Counter
import re  # For pattern matching

# API URL
url = "https://api.chess.com/pub/player/xreadytodayalex/games/2024/05"

# Send request with User-Agent header
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Function to fetch country name from the opponent's country URL
def fetch_country_name(country_url):
    response = requests.get(country_url, headers=headers)
    if response.status_code == 200:
        return response.json().get("name", "Unknown Country")
    return "Unknown Country"

# Function to fetch opponent details (including country)
def fetch_opponent_country(player_url):
    response = requests.get(player_url, headers=headers)
    if response.status_code == 200:
        player_data = response.json()
        country_url = player_data.get("country", "")
        if country_url:
            return fetch_country_name(country_url)
    return "Unknown Country"

# Function to clean opening/defense names
def clean_opening_name(opening_name):
    words = re.findall(r'\w+', opening_name)  # Extract words
    if len(words) >= 2:
        if words[0].lower() == "van" and words[1].lower() == "t":
            return "Van 't Kruijs"
        return ' '.join(words[:2])  # First two words
    return opening_name  # Fallback

# Main Script
response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()
    games = data.get("games", [])
    df = pd.DataFrame(games)

    # Filter for "rapid" and "rated" games only
    filtered_games = df[(df["time_class"] == "rapid") & (df["rated"] == True)].reset_index(drop=True)

    # Counters for wins, losses, draws, and openings/defenses
    wins = 0
    losses = 0
    draws = 0
    played_as_white = 0
    played_as_black = 0
    openings_defenses_used = Counter()

    # Process each game
    for i, row in filtered_games.iterrows():
        # Extract player usernames and results
        white_player = row['white']['username']
        black_player = row['black']['username']
        result_white = row['white']['result']
        result_black = row['black']['result']
        game_url = row['url']
        played_as = "White" if white_player == "XREADYTODAYALEX" else "Black"

        # Fetch opponent and country
        opponent_username = black_player if played_as == "White" else white_player
        opponent_id_url = row['black']['@id'] if played_as == "White" else row['white']['@id']
        opponent_country = fetch_opponent_country(opponent_id_url)

        # Extract and clean the opening/defense name
        raw_opening = row.get('eco', 'Unknown Opening').split("/")[-1]
        opening_name_cleaned = clean_opening_name(raw_opening)

        # Determine the winner
        if result_white == "win" and white_player == "XREADYTODAYALEX":
            winner = "XREADYTODAYALEX"
            wins += 1
        elif result_black == "win" and black_player == "XREADYTODAYALEX":
            winner = "XREADYTODAYALEX"
            wins += 1
        elif result_white in ["repetition", "agreed", "insufficient", "stalemate", "timevsinsufficient"] or \
             result_black in ["repetition", "agreed", "insufficient", "stalemate", "timevsinsufficient"]:
            winner = "Draw"
            draws += 1
        else:
            winner = opponent_username
            losses += 1

        # Count the opening/defense used
        openings_defenses_used[opening_name_cleaned] += 1

        # Track played as white or black
        if played_as == "White":
            played_as_white += 1
        else:
            played_as_black += 1

        # Print detailed game info
        print(f"Game {i+1}: {game_url}")
        print(f"   Played as: {played_as}")
        print(f"   Opponent: {opponent_username}")
        print(f"   Opponent's Country: {opponent_country}")
        print(f"   Opening/Defense: {opening_name_cleaned}")
        print(f"   Winner: {winner}\n")

    # Print summary results
    print("Summary Results:")
    print(f"Total Wins by XREADYTODAYALEX: {wins}")
    print(f"Total Losses by XREADYTODAYALEX: {losses}")
    print(f"Total Draws: {draws}")
    print(f"Total Games Played as White: {played_as_white}")
    print(f"Total Games Played as Black: {played_as_black}")

    # Print total number of openings/defenses used
    print("\nOpenings/Defenses Used by XREADYTODAYALEX:")
    for opening, count in openings_defenses_used.most_common():
        print(f"   {opening}: {count}")

else:
    print(f"Error occurred! HTTP Status Code: {response.status_code}")
```
